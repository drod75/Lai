```python
from datasets import load_dataset
import pandas as pd
```

```python
dataset = load_dataset("tingchih/Multi_News_fact_checking_claims", split="train")
df = dataset.with_format("pandas")
```

```python
df.replace(
    ["Entailment", "Neutral", "Contradiction"],
    ["True", "Nuanced", "False"],
    inplace=True,
)
df["Documents"] = df["Documents"].astype("string")
df["Documents"] = df["Documents"].str.replace("[", "")
df["Documents"] = df["Documents"].str.replace("]", "")
df["Documents"] = df["Documents"].str.replace("'", "")
df["Documents"].iloc[0]
```
