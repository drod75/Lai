Testing using the datasets library to merge data, specifically the Farsight dataset, and the Tingchih dataset

```python
from datasets import load_dataset
from dotenv import load_dotenv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

load_dotenv()
n_rows = 3000
```

```python
dataset_facts = load_dataset("Farsight-AI/10k-fact-check-finetune", split="train", streaming=True)
subset_facts = dataset_facts.take(n_rows)
facts_df = pd.DataFrame(list(subset_facts))

dataset_news = load_dataset("tingchih/Multi_News_fact_checking_claims", split="train", streaming=True)
subset_news = dataset_news.take(n_rows)
news_facts_df = pd.DataFrame(list(subset_news))
```

```python
facts_df['label'].value_counts()
```

```python
news_facts_df.info()
```

```python
news_facts_df['label'].value_counts()
```

```python
news_facts_df.replace(['Entailment', 'Neutral', 'Contradiction'], ['True', 'Nuanced', 'False'], inplace=True)
```

```python
facts_df.head()
```

```python
news_facts_df.head()
```

```python
news_facts_df['Documents'] = news_facts_df['Documents'].astype("string")
news_facts_df['Documents'] = news_facts_df['Documents'].str.replace('[', '')
news_facts_df['Documents'] = news_facts_df['Documents'].str.replace(']', '')
news_facts_df['Documents'] = news_facts_df['Documents'].str.replace("'", '')
news_facts_df['Documents'].iloc[0]
```

```python
news_facts_df.head()
```

```python
select_facts_df = facts_df[['context', 'answer', 'label']]
select_facts_df.columns = ['documents', 'claim', 'label']
news_facts_df.columns = ['documents', 'claim', 'label']
```

```python
select_facts_df.head()
```

```python
claims_df = pd.concat([select_facts_df, news_facts_df])
claims_df.info()
```

```python
claims_df.tail()
```

Merging these datasets can provide useful results, since they are malleable, and can be used to create a larger dataset for training a model.

## Respit dataset work

```python
dataset_repslit = load_dataset("jpd459/resplit_multi_fact_checking_dataset_resampling", split="train", streaming=True)
subset_resplit = dataset_repslit.take(4000) # the dataset has 3.3k rows
resplit_df = pd.DataFrame(list(subset_resplit))
print(resplit_df.info())
resplit_df.head()
```

```python
resplit_df['evidence_stance'].value_counts()
```

```python
def label_rework(x):
    if x == 'refutes' or x == 'insufficient-refutes':
        return 'False'
    elif x == 'supports' or x == 'insufficient-supports':
        return 'True'
    else:
        return 'Nuanced'
    
resplit_df['evidence_stance'] = resplit_df['evidence_stance'].apply(label_rework)
resplit_df.head()
```

```python
resplit_df.columns = ['claim', 'documents', 'label']
resplit_df.info()
```

```python
claims_df = pd.concat([claims_df, resplit_df])
claims_df.info()
```

8k rows, though more are possible with the full dataset of the tingchih/Multi_News_fact_checking_claims dataset. For finetuning, most likely that dataset will be used, as it contains more data.
